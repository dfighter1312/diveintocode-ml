**[Problem 1] Create using the exponentiation arithmetic operator**

```
"""
Code to calculate the thickness when the paper is folded once
"""
THICKNESS = 0.00008
folded_thickness = THICKNESS*(2**43)
print("Thickness: {} meters".format(folded_thickness))
```

**[Problem 2] Unit Conversion**

```
# Convert meters to 10,000 kilometers and display with two decimal places
print("Thickness: {: .2f} x 10,000 kilometers".format(folded_thickness/1000/10000))
```

The distance to the moon is 384400 km = 38.44 x 10000 km therefore the folded paper can reach the moon.

**[Problem 3] Create using a for statement**

```
mul = 1
exp = 43
for i in range(exp):
  mul *= 2
folded_thickness = THICKNESS * mul
print("Thickness: {} meters".format(folded_thickness))
```

**[Problem 4] Comparison of calculation time**

Check calculation speed of the first method.

```
%%timeit
THICKNESS = 0.00008
folded_thickness = THICKNESS*(2**43)
```

Check calculation speed of the second method.

```
%%timeit
mul = 1
exp = 43
for i in range(exp):
  mul *= 2
folded_thickness = THICKNESS * mul
```

As can clearly be seen, the first way of writing is better in time comparison (66.4 ns << 2.41 μs)

**[Problem 5] Saving to a list**

```
folded_thickness = THICKNESS
# The record of 43 folds, including the initial thickness as the first element
folded_lst = [folded_thickness]
exp = 43
for i in range(exp):
  folded_thickness *= 2
  folded_lst.append(folded_thickness)
# Confirm that the list contains 44 values and check the last value
len(folded_lst), folded_lst[-1]
```

**[Problem 6] Displaying a line graph**

```
"""
Display the graph. Title and axis label name.
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Thickness of folded paper")
plt.xlabel("Number of folds")
plt.ylabel("Thickness[m]")
plt.plot(folded_lst)
plt.show()
```

Since the thickness of n-folded paper is expressed as an exponential function, the thickness increases rapidly when n gets higher and higher.

**[Problem 7] Customizing graphs**

```
"""
Display a graph with green circle marker.
"""
plt.title("Thickness of folded paper")
plt.xlabel("Number of folds")
plt.ylabel("Thickness[m]")
plt.plot(folded_lst, 'go')
plt.show()
```

```
"""
Display a graph with red plus and adjust line width and marker size.
"""
plt.title("Thickness of folded paper")
plt.xlabel("Number of folds")
plt.ylabel("Thickness[m]")
plt.plot(folded_lst, 'r+', linewidth=2, markersize=12)
plt.show()
```

```
"""
Display a graph with blue dashed line.
"""
plt.title("Thickness of folded paper")
plt.xlabel("Number of folds")
plt.ylabel("Thickness[m]")
plt.plot(folded_lst, marker='o', linestyle=':', color='darkgrey', fillstyle='bottom', markerfacecolor='blue')
plt.show()
```
