## Assignment 2 - Creating Dummy Data
**[Problem 1] Creating random numbers**

```python
import numpy as np
mean = np.array([-3, 0])
cov = np.array([[1.0, 0.8], [0.8, 1.0]])
random_ndarray = np.random.multivariate_normal(mean, cov, 500)
#Check the shape of the newly created array
print(random_ndarray.shape)
```

**[Problem 2] Visualization by scatter plot**

```python
# Store x and y value seperately for all data points
x, y = random_ndarray.T

import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.show()
```

**[Problem 3] Visualization by histogram**

```python
plt.title("Histogram of x")
plt.xlim(left=-6, right=6)
plt.hist(x)
plt.show()
plt.title("Histogram of y")
plt.hist(y)
plt.xlim(left=-6, right=6)
plt.show()
```

**[Problem 4] Addition of data**

```python
mean_a = np.array([0, -3])
cov_a = np.array([[1.0, 0.8], [0.8, 1.0]])
random_ndarray_a = np.random.multivariate_normal(mean_a, cov_a, 500)
x_a, y_a = random_ndarray_a.T
plt.scatter(x, y, label='0')
plt.scatter(x_a, y_a, c='m', label='1')
plt.legend()
plt.show()
```

**[Problem 5] Data combination**

```python
#Data combination using np.concatenate
combined_ndarray = np.concatenate((random_ndarray, random_ndarray_a))
#Check the dimension of the combined array
print(combined_ndarray.shape)
```

```python
#Data combination using np.vstack
another_combined_ndarray = np.vstack((random_ndarray, random_ndarray_a))
print(another_combined_ndarray.shape)
```

**[Problem 6] Labeling**

```python
#Create an array with 500 consecutive 0s and 500 consecutive 1s
label_ndarray = np.concatenate((np.zeros(500), np.ones(500))).reshape(1000,1)

#Combine the array from problem 5 to the label array
#It can be done by using np.concatenate with axis=1 or np.hstack (to stack the arrays vertically)
labelled_ndarray = np.concatenate((combined_ndarray, label_ndarray), axis=1)

#Check the shape of the newly created array
print(labelled_ndarray.shape)

#Check if the elements is correctly labelled
print(labelled_ndarray[:10])
print(labelled_ndarray[495:505])
print(labelled_ndarray[-10:])
```
