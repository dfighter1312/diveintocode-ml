**[Problem 1] Number of grains of rice on the 100th day**

```python
# Number of grains of rice on the 4th day (for testing)
grain_first_day = 1
day = 4    

# List to store and display in a graph
grain_total = [grain_first_day]
grain_get = grain_total[:]

for i in range(day - 1):
    grain_get.append(grain_get[-1] * 2)
    grain_total.append(grain_get[-1] + grain_total[-1])
    
# Display the graph
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Number of grains get in a day and total")
plt.xlabel("Day")
plt.ylabel("Number of grains")
plt.plot(range(1, day + 1), grain_get, label="1 day")
plt.plot(range(1, day + 1), grain_total, 'y', label="total")
plt.legend()
plt.show()
```

**[Problem 2] Number of rice grains outside of the 100th day**

```python
def compute_sorori_shinzaemon(day=100, grain=1):
    """
    A function that verifies the anecdote of rice by Sorori Shinzaemon. Calculate the number of rice grains per day.
    Parameteres
    --------------
    day: int
        Number of days to get rice (default : 100)
    Returns
    --------------
    list_n_grains : list
        List of the number of rice you get one day (the length of the list is equal to day)
    list_total_grains : list
        List of the total number of rice you will receive by a certain day (the length of the list will be equal to day)
    """
    list_n_grains = [grain]
    list_total_grains = list_n_grains[:]
    for i in range(day - 1):
        list_n_grains.append(list_n_grains[-1] * 2)
        list_total_grains.append(list_n_grains[-1] + list_total_grains[-1])
    return list_n_grains, list_total_grains
```

```python
# Test with day = 4
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=4)
print("You will get {} grains of rice in day 4, and get total of {} in that day.".format(list_n_grains[-1], list_total_grains[-1]))

# Test with day = 100
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=100)
print("You will get {} grains of rice in day 100, and get total of {} in that day.".format(list_n_grains[-1], list_total_grains[-1]))
```

**[Problem 3] How many people can live for how many days with the rice we get?**

*It takes 400 grams or about 19,200 grains of rice to feed one adult for a day.* (via [Wikipedia](https://en.wikipedia.org/wiki/Freerice))

We will take the result of number of grains of rice after a certain day from the function created in problem 2. Since we only need the total number of grains of rice in the last day, we have the grains variable equal to <span style="color:blue">compute_sorori_shinzaemon(day)[1][-1]</span>. The first index [1] tells us to access list_total_grains and the second index [-1] tells us to get the last element of the list.

```python
def compute_days_to_live(day, people):
    """
    Calculate how many people can live for how many days with this rice.
    Parameters:
    ------------------
    day: int
        Number of days to get rice
    people: int
        Number of people need to be computed
    Returns:
    ------------------
    day_to_live: float
        Number of days people can live
    """
    grains = compute_sorori_shinzaemon(day=day)[1][-1]
    day_to_live = grains / (19200 * people)
    return day_to_live
```

```python
#Testing the function
dtl1 = compute_days_to_live(16, 1)
print("A person can live in {:.1f} days with the number of grains of rice after 16 days.".format(dtl1))
dtl = compute_days_to_live(50, 50000)
print("50,000 people can live in {:.1f} days with the number of grains of rice after 50 days.".format(dtl))
```
