### [Problem 1] Data acquisition

```python
from sklearn.datasets import load_iris
import pandas as pd

#Get an overview of the data
load_iris()
```

```python
#Store the explanatory variables in the Pandas DataFrame type variable X
X = pd.DataFrame(data=load_iris().data, columns=['sepal_length','sepal_width','petal_length','petal_width'])
X
```

```python
#Store the objective variable in y
y = pd.DataFrame(load_iris().target, columns=[('Species')])
y
```

### [Problem 2] Combining data

```python
df = pd.concat([X, y], axis=1)
df
```

### [Problem 4] Examining the iris dataset itself

- It is a multivariate data set introduced by the British statistician, eugenicist, and biologist Ronald Fisher in his 1936 paper The use of multiple measurements in taxonomic problems as an example of linear discriminant analysis.
- The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters.
- Based on Fisher's linear discriminant model, this data set became a typical test case for many statistical classification techniques in machine learning such as support vector machines.

### [Problem 6] Creating a diagram

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Make a pie chart of the number of samples per label (while showing percentages).

```python
sr = df["Species"].value_counts()
sr.plot.pie(autopct="%.1f%%");
```

Select features one by one and visualize the distribution of data for each label using a box plot.

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 18))

fig.suptitle('Distribution of data for each label using a box plot')

sns.boxplot(ax=ax[0,0], x='Species', y='sepal_length', data=df)
sns.boxplot(ax=ax[0,1], x='Species', y='sepal_width', data=df)
sns.boxplot(ax=ax[1,0], x='Species', y='petal_length', data=df)
sns.boxplot(ax=ax[1,1], x='Species', y='petal_width', data=df)

sns.set()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(18, 18))

fig.suptitle('Distribution of data for each label using a violin plot')

sns.violinplot(ax=ax[0,0], x='Species', y='sepal_length', data=df)
sns.violinplot(ax=ax[0,1], x='Species', y='sepal_width', data=df)
sns.violinplot(ax=ax[1,0], x='Species', y='petal_length', data=df)
sns.violinplot(ax=ax[1,1], x='Species', y='petal_width', data=df)

sns.set()
```

**Box plot** only shows summary statistics such as mean/median and interquartile ranges.

**The violin plot** shows the full distribution of the data.

### [Problem 7] Confirming the relationship between features

```python
fig, ax = plt.subplots(2, 3, figsize=(15,10))

fig.suptitle('Scatter plot')

sns.scatterplot(ax=ax[0,0], data=df, x='sepal_length', y='sepal_width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[0,1], data=df, x='sepal_length', y='petal_length', hue='Species', palette="deep")
sns.scatterplot(ax=ax[0,2], data=df, x='sepal_length', y='petal_width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1,0], data=df, x='sepal_width', y='petal_length', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1,1], data=df, x='sepal_width', y='petal_width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1,2], data=df, x='petal_length', y='petal_width', hue='Species', palette="deep")

sns.set()
```
