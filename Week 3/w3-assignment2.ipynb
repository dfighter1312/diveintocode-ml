## Assignment 2 - A question of going down Mt. Fuji

**Get the elevation data**

The file is got via https://en.wikipedia.org/wiki/List_of_mountains_and_hills_of_Japan_by_height. I had converted all tables in this site to .csv files and named it *mtfuji_data.csv*

```python
import numpy as np

csv_path = "mtfuji_data.csv"
np.set_printoptions(suppress=True)
fuji = np.loadtxt(csv_path, delimiter=',', skiprows=1)

#Print sample data
print(fuji[25:30])
```

**[Problem 1] Visualization of data**

```python
tranpose_fuji = fuji.T

from matplotlib import pyplot as plt
%matplotlib inline
plt.title('Mt.Fuji')
plt.xlabel('position')
plt.ylabel('elevation [m]')
plt.plot(tranpose_fuji[0], tranpose_fuji[3])
plt.show()
```

**[Problem 2] Create a function to calculate the slope of a point**

```python
def get_slope(array_x, array_y):    
    array_xy = np.column_stack((array_x, array_y))
    diff = (array_xy[1:] - array_xy[:-1]).T
    slope = diff[1] / diff[0]
    return slope

slope = get_slope(tranpose_fuji[0], tranpose_fuji[3])
```

**[Problem 3] Create a function to calculate the destination**

```python
def get_next_dest(current_loc, gradient_loc, alpha=0.2):
    return current_loc - alpha * gradient_loc

print(tranpose_fuji.shape)
next_dest = get_next_dest(tranpose_fuji[3][:-1], slope)
```

```python
print(next_dest.shape, tranpose_fuji[3][:-1].shape)
print(np.column_stack((next_dest, tranpose_fuji[3][:-1])))
```
