# Mathematics lesson for Machine Learning
## Assignment 2 - A question of going down Mt. Fuji

**Get the elevation data**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

csv_path = "mtfuji_data.csv"
np.set_printoptions(suppress=True)
fuji = np.loadtxt(csv_path, delimiter=',', skiprows=1)

#Print sample data
print(fuji[25:30])
```

**[Problem 1] Visualization of data**

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.title('Mt.Fuji')
plt.xlabel('position')
plt.ylabel('elevation [m]')
plt.plot(fuji[:,0], fuji[:,3])
plt.show()
```

**[Problem 2] Create a function to calculate the slope of a point**

```python
def get_slope(array_x, array_y):    
    array_xy = np.column_stack((array_x, array_y))
    diff = (array_xy[1:] - array_xy[:-1]).T
    slope = diff[1] / diff[0]
    return slope

slope = get_slope(fuji[:, 0], fuji[:, 3])
print(slope[:10])
```

**[Problem 3] Create a function to calculate the destination**

```python
def get_next_dest(current_loc, gradient_loc, min_val, max_val, alpha=0.2):
    next_dest = current_loc - alpha * gradient_loc
    if next_dest < min_val or next_dest > max_val:
        raise ValueError("The destination point exceeds its range.")
    return round(next_dest)

#Testing the function
point = 112
min_val = fuji.T[0].min()
max_val = fuji.T[0].max()
next_dest = get_next_dest(fuji[:,0][point], slope[point], min_val, max_val)
print(next_dest)
```

**[Problem 4] Create a function that goes down the mountain**

```python
def go_down_the_mountain(point_array, slope_array, first_loc, alpha=0.2):
    min_val = point_array.min()
    max_val = point_array.max()
    visited = []
    point_vst = first_loc
    while point_vst not in visited:
        visited.append(point_vst)
        point_vst = get_next_dest(point_array[point_vst], slope_array[point_vst], min_val, max_val, alpha)
    return visited

first_loc = 136
result = go_down_the_mountain(fuji[:,0][:-1], slope, first_loc)
print(result)
```

```python
descended_elevations = fuji[:,3][result]

plt.title('Downhill process of Mt. Fuji')
plt.ylabel('elevation[m]')
plt.xlabel('position')
plt.plot(fuji[:,0], fuji[:,3])
plt.scatter(result, descended_elevations, 20, color='red')
plt.show()
```

```python
# Make a line graph of the elevation at each iteration.
plt.title('Downhill process of Mt. Fuji')
plt.ylabel('elevation[m]')
plt.xlabel('number of iterations')
plt.plot(descended_elevations)
plt.show()
```

**[Problem 6] Change of initial value**

```python
def draw_downhill_process(result, text=''):
    descended_elevations = fuji[:,3][result]
    
    # Make a line graph of downhill process.
    plt.title('Downhill process of Mt. Fuji ' + text)
    plt.ylabel('elevation[m]')
    plt.xlabel('position')
    plt.plot(fuji[:,0], fuji[:,3])
    plt.scatter(result, descended_elevations, 20, color='red')
    plt.show()

    # Make a line graph of the elevation at each iteration.
    plt.title('Downhill process of Mt. Fuji')
    plt.ylabel('elevation[m]')
    plt.xlabel('number of iterations')
    plt.plot(descended_elevations)
    plt.show()

first_loc = 142
result = go_down_the_mountain(fuji[:, 0][:-1], slope, first_loc)
print(result)
draw_downhill_process(result)
```

```python
first_loc = 220
result = go_down_the_mountain(fuji[:, 0][:-1], slope, first_loc)
print(result)
draw_downhill_process(result)
```

**[Problem 7] (Advance problem) Change hyperparameters**

```python
first_loc = 136
for alpha in np.arange(0.1, 0.5, 0.1):
    result = go_down_the_mountain(fuji[:, 0][:-1], slope, first_loc, alpha)
    draw_downhill_process(result, text='(alpha = ' + str(round(alpha, 2)) + ')')
```
